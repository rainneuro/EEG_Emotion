```python
%matplotlib
import os
import cv2
import numpy as np
from PIL import Image
import mne
from mne.viz import plot_topomap
from mne.time_frequency import tfr_morlet, psd_multitaper, psd_welch, psd_array_multitaper
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Use the mne.read_epochs_eeglab() to read the preprocessed data
demo = mne.read_epochs_eeglab('F:/1_Emotion_Data/EEG/Angry/sub_001_ag1.set')
montage = mne.channels.read_custom_montage('F:/1_Emotion_Data/EEG/Emotion.loc')
# Apply the montage file to the demo data
demo.set_montage(montage)
```

```python
# Create the frequency-band index
delta = freqs[0:4];theta = freqs[3:7]; alpha = freqs[7:13]; beta = freqs[13:30]; gamma = freqs[30:45]; 
```

```python
delta_dir = 'F:/1_Emotion_Data/EEG/Topos/Delta/'; theta_dir = 'F:/1_Emotion_Data/EEG/Topos/Theta/'; alpha_dir = 'F:/1_Emotion_Data/EEG/Topos/Alpha/'; beta_dir = 'F:/1_Emotion_Data/EEG/Topos/Beta/'; gamma_dir = 'F:/1_Emotion_Data/EEG/Topos/Gamma/'

for i in range(0,274):
    epoch = demo[i]
    psds, freqs = psd_multitaper(epoch,fmin=1, fmax=50)
    psds = 10* np.log10(10**12*psds)
    
    delta = np.average(np.squeeze(psds)[:,0:4],axis=1)
    plot_topomap(delta, demo.info, sphere= 0.13)
    delta_file = os.path.join(delta_dir, str(i) + '.png')
    plt.savefig(delta_file,bbox_inches='tight',dpi=600,pad_inches=0.1)
    plt.close()
    
    theta = np.average(np.squeeze(psds)[:,3:7],axis=1)
    plot_topomap(theta, demo.info, sphere= 0.13)
    theta_file = os.path.join(theta_dir, str(i) + '.png')
    plt.savefig(theta_file,bbox_inches='tight',dpi=600,pad_inches=0.1)
    plt.close()
    
    alpha = np.average(np.squeeze(psds)[:,7:13],axis=1)
    plot_topomap(alpha, demo.info, sphere= 0.13)
    alpha_file = os.path.join(alpha_dir, str(i) + '.png')
    plt.savefig(alpha_file,bbox_inches='tight',dpi=600,pad_inches=0.1)
    plt.close()
    
    beta = np.average(np.squeeze(psds)[:,13:30],axis=1)
    plot_topomap(beta, demo.info, sphere= 0.13)
    beta_file = os.path.join(beta_dir, str(i) + '.png')
    plt.savefig(beta_file,bbox_inches='tight',dpi=600,pad_inches=0.1)
    plt.close()
    
    gamma = np.average(np.squeeze(psds)[:,30:45],axis=1)
    plot_topomap(gamma, demo.info, sphere= 0.13)
    gamma_file = os.path.join(gamma_dir, str(i) + '.png')
    plt.savefig(gamma_file,bbox_inches='tight',dpi=600,pad_inches=0.1)
    plt.close()
```

```python
def delta_video(video_path,fps):
   # path: video output path fps: frame per second per second
    fourcc = cv2.VideoWriter_fourcc(*"mp4v")
    # Decalre the path of all the topoplots
    topo_dir = 'F:/1_Emotion_Data/EEG/Topos/Delta' 
    im = Image.open('F:/1_Emotion_Data/EEG/Topos/Delta/0.png')  # 其中一张图像
    print(im.size)
    vw = cv2.VideoWriter(video_path, fourcc, fps, im.size)
    for i in os.listdir(topo_dir):
        frame = cv2.imread(topo_dir + '/' + i)
        vw.write(frame)

# Define our video path
video_path = 'F:/1_Emotion_Data/EEG/Topos/Videos/Delta.mp4'
# Convert the topoamps into video with a fps of 
delta_video(video_path,20) 
```
